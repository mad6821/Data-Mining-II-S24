```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import KernelPCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('dalton_df.csv')
df_small = df[["oindex", "new_oindex", "v2x_polyarchy"]]
```

### Correlation Analysis - Raw Data

```python
# Correlation matrix
corr_mat = df_small.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_mat, annot=True, cmap='coolwarm', vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.title('Correlation Plot')
plt.show()
```

### Kernel PCA

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_small.values.reshape(-1, 1))

# Apply Kernel PCA 
kpca = KernelPCA(n_components=1, kernel='rbf')
X_kpca = kpca.fit_transform(df_scaled)

# Extract PCA Scores
pca_scores = pd.DataFrame(X_kpca)
```

```python
plt.title("Kernel PCA")
plt.hist(X_kpca)
#plt.scatter(X_kpca[:, 0], X_kpca[:, 1])
plt.show()
```
